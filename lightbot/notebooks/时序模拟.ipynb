```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2022)
num = 24 * 60 * 60 * 365
x = np.random.choice([-1, 1], num) * 0.01
y = np.cumsum(x)
y = y + (-y.min()) + 3
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(y[::1000])
plt.show()
```
